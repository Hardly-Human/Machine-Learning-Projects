# support vector machine

```python
import pandas as pd

data = pd.read_csv('data.csv',names = ['x1','x2','y'])
print(data.shape)
data.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['x1'],data['x2'],c = data['y'])
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(data[['x1','x2']], data['y'] , test_size = 0.1 )
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
plt.figure(figsize = (12,3))

plt.subplot(1,2,1)
plt.scatter(x = x_train['x1'] , y = x_train['x2'], c = y_train)

plt.subplot(1,2,2)
plt.scatter(x = x_test['x1'], y = x_test['x2'], c = y_test)
```

```python
from sklearn.svm import SVC

model = SVC(kernel = 'rbf', gamma = 27)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score

accuracy = accuracy_score(y_test, y_pred)
precision  = precision_score(y_test , y_pred)

print('accuracy :' , accuracy)
print('precision :' ,precision)
```

```python
print('y_pred : ', y_pred)
print('y_true : ', list(y_test))
```
